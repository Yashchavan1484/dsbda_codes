```python
import pandas as pd
import numpy as np
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df = df.drop('Cabin', axis=1)
```

```python
df.isnull().sum()
```

```python
df_encoded = pd.get_dummies(df, columns=['Sex', 'Embarked'])
```
