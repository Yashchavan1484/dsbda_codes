```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('titanic')
dataset
```

```python
tc = dataset.corr(numeric_only=True)
sns.heatmap(tc, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=dataset)
plt.title('Survival Count by Ticket Class')
plt.show()
```

```python
sns.barplot(x='sex', y='survived', data=dataset)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.histplot(dataset, x='fare', bins=30)
plt.title('Histogram of Fare')
plt.show()
```

```python
sns.distplot(dataset['fare'])
plt.title('Distplot of Fare')
plt.show()
```

```python
sns.distplot(dataset['fare'],kde=False,bins=5)
```

```python
sns.distplot(dataset['fare'],hist=False)
```

```python
sns.jointplot(x='age',y='fare',data=dataset)
```

```python
sns.rugplot(x=dataset['fare'])
plt.title('Rug Plot of Fare')
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=dataset)
plt.title('Barplot of Fare vs Class')
plt.show()
```
